```python
import random
# this game follows this path: seer sees a good person marked as gold, king werewolf reveals and kills seer leaving the good with a gold
# witch save herself on first night or the person died on the second night
# Hunter does use his gun with a rate of good 30% except the gold
# Guardian guards himself first round and next the gold next himself if the gold is killed, then randomly 30% of good choice
# if there is a silver, guard the silver as next choice
# Killing and expelling are random but expulsion not gold neither silver

wins = []
for i in range(0,100000):
    # initialization: generate the roles and assign indices
    alive = set(range(1,13))
    king = set(random.sample(range(1, 13),1))
    ww = set(random.sample(list(alive - king), 3))
    bad = king | ww 
    good = alive - ww - king
    hunter = {random.choice(list(good))}
    witch = {random.choice(list(good - hunter))}
    guardian = {random.choice(list(good - witch - hunter))}
    seer = {random.choice(list(good - witch - hunter - guardian))}
    v = good - witch - hunter - guardian - seer
    functional = good - v
    potion_s = 1
    potion_k = 1
    kill = set()
    dead = set()
    wanted = set()
    expel = set()
    kill_tn = set()
    save = 0
    poison = 0
    kill_w = set()
    coin = 0
    checked = set()
    day = 0
    win = 0
    silver = set()
    kill_h = set()
    gold = set()

    # game starts
    while win == 0:

        # the night falls, the bad kills a person
        if set.intersection(bad,alive) != set():
            kill_tn = {random.choice(list(good - guardian))}

        # seer checks id
        if set.intersection(seer,alive) != set():
            gold = {random.choice(list(good - seer))}


        # if kill == witch, the witch will save herself
        if set.intersection(witch,alive) != set():
            if potion_s == 1 and day == 1:
                save = 1
                potion_s = 0
                silver = kill_tn
                
            if wanted != set() and potion_k == 1:
                kill_w = wanted
                potion_k = 0
                poison = 1
                

        # the guardian choose a person to guard:
        if set.intersection(guardian,alive) != set():
            if day % 2 == 0:
                guard = guardian
            elif day == 1 :
                guard = {0}
            elif gold != set() and set.intersection(gold,alive) != set():
                guard = gold
            elif save == 0 and silver != set() and set.intersection(gold,alive) == set():
                guard = silver
            elif random.uniform(0,1) < 0.3:
                guard = {random.choice(list(set.intersection(alive,good)))}
            else : guard = {random.choice(list(alive))}
            if guard == silver and save == 1 :
                coin = 1    

        # sun comes out, on day 1 the president election run and refresh alive 
        day = day + 1

        if coin == 1:
            kill = kill | kill_tn | kill_w
            coin = 0
            save = 0
        elif save == 1:
            kill = kill
            save == 0
        else:
            kill = kill | kill_tn | kill_w
            
        poison = 0

        if day == 1:
            # the king gets up and takes down the seer 
            dead = king | seer | kill
            # alive changes
        else: dead = dead | kill

        alive = alive - dead

        if set.intersection(bad,alive) == set():
            win = 1
            break
        elif set.intersection(v,alive) == set() or set.intersection(functional,alive) == set():
            win = -1
            break
            
        # the hunter shows 
        if set.intersection(hunter,alive) == set() and kill_h == set():
            if random.uniform(0,1) < 0.3:
                kill_h = {random.choice(list(set.intersection(bad,alive)))}
            elif random.uniform(0,1) < 0.5:
                kill_h = {random.choice(list(alive-gold))}
            else :
                kill_h = {0}
            dead = dead | kill_h
            alive = alive - dead
            
        if set.intersection(bad,alive) == set():
            win = 1
            break
        elif set.intersection(v,alive) == set() or set.intersection(functional,alive) == set():
            win = -1 
            break

        if wanted != set() and set.intersection(wanted,alive) != set():
            expel = wanted
        elif set.intersection(witch,alive) != set() :
            expel = {random.choice(list(alive-silver-gold))}
        else :
            expel = {random.choice(list(alive-gold))}

        dead = dead | expel

        alive = alive - dead

        if set.intersection(bad,alive) == set():
            win = 1
            break
        elif set.intersection(v,alive) == set() or set.intersection(functional,alive) == set():
            win = -1 
            break

        if set.intersection(wanted,dead) != set():
            wanted = set()
            
                # the hunter shows 
        if set.intersection(hunter,alive) == set() and kill_h == set():
            if random.uniform(0,1) < 0.3:
                kill_h = {random.choice(list(set.intersection(bad,alive)))}
            elif random.uniform(0,1) < 0.5:
                kill_h = {random.choice(list(alive-gold))}
            else :
                kill_h = {0}
            dead = dead | kill_h
            alive = alive - dead

    wins.append(win)

print "the good will win {} times out of 100000 games".format(wins.count(1))
print "the bad will win {} times out of 100000 games".format(wins.count(-1))
```
